```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from statistics import mean
from sklearn.linear_model  import LinearRegression
```

***Task 1***

```python
income = np.array([93.62,91.54,93.25,94.164,96.12,102.89,105.27,108.963,112.256,119.32,125.89,127.914])
```

```python
income_without_tax = income * .7
print(f'Income without tax is: {income_without_tax}')
```

```python
expenses = np.array([56.02, 56.75, 57.662, 59.25, 62.56, 64.094, 70.44, 73.68, 77.209, 82.89, 87.64, 88.884])
```

```python
savings = income_without_tax - expenses
print(savings)
```

```python
fin_res = pd.DataFrame(
    data = {'Income_without_tax': income_without_tax, 'Expenses' : expenses, 'Savings': savings})
fin_res['Month'] = ['January', 'February', 'March', 'April', 'May', 'June', 'July','August','September', 'October', 'November', 'December' ]
```

```python
months = {
    'January' : 1,
    'February' : 1,
    'March' : 1,
    'April' : 2,
    'May' : 2,
    'June' : 2,
    'July' : 3,
    'August' : 3,
    'September' : 3,
    'October' : 4,
    'November' : 4,
    'December' : 4
}
fin_res['Quarter'] = fin_res['Month'].map(months)
```

```python
plt.figure(figsize = (9,4))
bars = plt.bar(fin_res['Month'], fin_res['Savings'], width = .7, label = 'Savings' , align = 'center', color = 'yellow')
mplcyberpunk.add_bar_gradient(bars = bars)
plt.xticks(rotation = 45, size = 12)
plt.yticks( size = 12)
plt.title('Savings by month', size = 14)
plt.show()
```

```python
import mplcyberpunk
plt.style.use('cyberpunk')
plt.figure(figsize = (9,4))
plt.plot(fin_res['Month'], fin_res['Income_without_tax'], label = 'Income without tax', marker = 'o')
plt.grid(True)
plt.xticks(rotation = 45, size = 12)
plt.yticks( size = 12)
plt.title('Income after tax by month', size = 14)
mplcyberpunk.add_glow_effects()
plt.show()
```

```python
col = sns.color_palette('icefire')
plt.style.use('default')
plt.figure(figsize = (10, 6))
plt.pie(fin_res['Savings'], labels = fin_res['Month'], autopct = '%.2f%%',
       explode = [0.1, 0, 0, 0, 0, 0.12, 0, 0, 0, 0, 0, 0], colors = col )
plt.title('Savings by month')
plt.show()
```

```python
aver_income = fin_res.groupby('Quarter')['Income_without_tax'].mean().reset_index()
```

```python
aver_income['Qs'] = aver_income['Quarter'].apply(lambda x: 'Q' + str(x))
aver_income
```

```python
plt.figure(figsize = (9, 5))
plt.bar(aver_income['Qs'],  aver_income['Income_without_tax'], color = 'maroon', alpha = 0.6, width = .6)
plt.title('Income by quarter', size = 14)
plt.show()
```

***Task 2***

```python
car_crashes = sns.load_dataset('car_crashes')
```

```python
car_crashes.head()
```

```python
cc_heat = car_crashes.iloc[:, :7]
mask = np.triu(np.ones_like(cc_heat.corr()))
```

```python
plt.figure(figsize = (9,5))
corr_map = sns.heatmap(cc_heat.corr(), annot = True, fmt = '.2f', mask = mask)
corr_map.get_xticklabels()[0].set_color('red')
plt.title('Car crashes heatmap')
plt.show()
```

```python
car_crashes['previous'] = car_crashes['total'] - car_crashes['no_previous']
data = {
    'variables' : ['no_previous', 'previous'],
    'values' : [car_crashes['no_previous'].mean(),  car_crashes['previous'].mean() ]
}
expl = pd.DataFrame(data)
expl
```

```python
sns.barplot(expl, x= 'variables', y ='values')
plt.ylabel('%')
plt.xlabel('')
plt.show()
```

```python
cmap = sns.color_palette('husl', 8)
plt.pie(expl['values'], labels = expl['variables'], autopct = '%.2f%%', colors = cmap)
plt.title('Previously crashed and not crashed drivers as percentage of total accidents')
plt.show()
```

```python
sns.regplot(data = car_crashes, x = 'previous', y = 'total')
# plt.text(1, 60, r2_score(car_crashes['previous'],car_crashes['total']))
plt.title('Regression of previous crashes on total number of crashes')
plt.show()
```

```python
sns.regplot(data = car_crashes, x = 'alcohol', y = 'total')
plt.title('Regression of percentege of drunk drivers involved in fatal accidents to total accidents')
plt.show()
```

```python
sns.regplot(data = car_crashes, x = 'speeding', y = 'total')
plt.title('Regression of percentege of speeding drivers involved in fatal accidents to total accidents')
plt.show()
```

```python
plt.figure(figsize = (6,5))
sns.histplot(car_crashes['total'])
plt.title('Distribution of number of drivers involved in fatal car crashes by state')
plt.ylabel('Frequency')
plt.xlabel('Total crashes')
plt.show()
```

```python
top_5 = car_crashes.sort_values(by = 'total', ascending = False).head(5).reset_index(drop = True)
low_5 = car_crashes.sort_values(by = 'total', ascending = True).head(5).reset_index(drop = True)
top = pd.concat([low_5, top_5]).reset_index(drop = True)
```

```python
top.columns
```

```python
plt.figure(figsize = (9,5))
colors = ['green', 'green', 'green', 'green', 'green', 
         'red', 'red', 'red', 'red', 'red']
bars = sns.barplot(data = top, x = 'abbrev', y = 'total',palette = colors)
plt.xlabel('States')
plt.ylabel('Number of fatal car crashes')
plt.title('Top 5 states with highest and lowest number of fatal car accidents')
plt.show()
```
